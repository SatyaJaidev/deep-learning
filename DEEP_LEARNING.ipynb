**GRADIENT BASED OPTIMIZATION**

**STEP 1:** initially the "w" is a weight are filled with small random values, a step called random initialization.||               
**STEP 2:** training is the gradual adjustment of the weights such that we keep updating the weights so that the output matches the actual output.                                                                         
**STEP 3:** (TRAINING LOOP):suppose we have a bunch of images in our database in which few represent the image of a dog and others represent a cat,[(consider 1=dog,0=cat)],now, we draw a random samples,say we drew 4 samples from it which is going to be our "x", and the one which are selected with naming is going to be our "y",[(ex: say, we selected, (1,0,0,1)=y,and since there are four so x=4)]                                                                         || **STEP 4 :(FORWARD PASS)**:now we put the input x=4 in the formula of output and the whole solve problem is going to be our y_pred. ||
**STEP 5:** compute the whole loss of a batch and measure the mismatch between the y and y_pred,here,(y is true value and y_pred is predicted). ||
**STEP 6:(BACKWARD PASS)** we compute the graident of the loss with regard to the network, so basically we have to calculate how bad the loss is, when we say graident of the loss, what we mean is, say, there is a loss curve, say this is what the loss we got,[(refer the image)] the gradient tells us how deep to go or how much we should we move so that we can eventually get to the minimum loss possible, now , the point of calculating the gradient loss is so that we can update the weights. ||
**STEP 7:** move the parameters a little in the oppsoite direction from the gradient and follow the formula
**[W-=step*gradient]**.
so if y_pred is really good then no need of any update in the W but if y_pred is really bad then you have to change a huge margin in W. So we repeat the cycle over and over again until out W stabilizes. So hence this how we perform our gradient based optimization.
